```python

# Extracting data
country = covid_data['country']
deaths = covid_data['deaths']

# Creating the plot
plt.figure(figsize=(16, 8))
plt.plot(country, deaths, label="Deaths",  color='royalblue') 

# Adding labels and title
plt.xlabel('Country')
plt.ylabel('Number of Deaths')
plt.title('COVID-19 Deaths by Country')
plt.grid()

# Adding legend
plt.legend()

# Display the plot
plt.show()
```

```python
# groupby sum of confirmed case 
country_confirmed=covid_data.groupby('country')['confirmed'].sum()
country_confirmed.to_frame()
```

```python
# extracting top 15 country's data
top_15_confirmed=covid_data.sort_values(by='confirmed',ascending=False).head(15)
top_15_confirmed
```

```python
# Extracting data
country = top_15_confirmed['country']
confirmed = top_15_confirmed['confirmed']

# creating bar plot 
plt.figure(figsize=(12, 8))
plt.bar(country, confirmed, label="confirmed cases ", align='edge',  width=0.3, color='royalblue', edgecolor='black') 

# Adding labels and title
plt.xlabel('top 15 country')
plt.ylabel('Number of confirmed case')
plt.grid()
plt.title('COVID-19 confirmed cases of Country')

# Adding legend
plt.legend()

# display plot 
plt.show()
```

```python
# top 15 country's critical cases data
```

```python
country_crtical=covid_data.groupby('country')['critical'].sum()
country_crtical.to_frame()
```

```python
# extracting top 15 crtical cases conutry's data 
top_15_critical=covid_data.sort_values(by='critical',ascending=False).head(15)
top_15_critical
```

```python
# Extracting data
country = top_15_critical['country']
critical = top_15_critical['critical']

# creating scatter plot 
plt.figure(figsize=(18, 8))
plt.scatter(country, critical , label="critical cases ",color='green', edgecolor='black',marker = 'D',s=80) 

# Adding labels and title
plt.xlabel('top 15 country')
plt.ylabel('Number of critical case')
plt.grid()
plt.title('COVID-19 critical cases of Country')

# Adding legend
plt.legend()

# display plot 
plt.show()
```
